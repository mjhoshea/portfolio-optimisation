```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import random 

plt.style.use(['science', 'retro', 'grid'])
```

```python
mean = 0
standard_deviation = np.sqrt(0.2)

mean2 = 0
standard_deviation2 = 1

mean3 = 0
standard_deviation3 = np.sqrt(5)

mean4 = -2
standard_deviation4 = np.sqrt(0.5)

mean5 = 3
standard_deviation5 = np.sqrt(0.25)

x_values = np.arange(-5, 5, 0.01)

y_values_1 = stats.norm(mean, standard_deviation)
y_values_2 = stats.norm(mean2, standard_deviation2)
y_values_3 = stats.norm(mean3, standard_deviation3)
y_values_4 = stats.norm(mean4, standard_deviation4)
y_values_5 = stats.norm(mean5, standard_deviation5)




plt.figure(figsize=(8, 6), dpi=100)
plt.plot(x_values, y_values_1.pdf(x_values), label=r'$\mu=0, \,\;\;\, \sigma^2=0.2$')
plt.plot(x_values, y_values_2.pdf(x_values), label=r'$\mu=0, \,\;\;\, \sigma^2=1$')
plt.plot(x_values, y_values_3.pdf(x_values), label=r'$\mu=0, \,\;\;\, \sigma^2=5$')
plt.plot(x_values, y_values_4.pdf(x_values), label=r'$\mu=-2, \, \sigma^2=0.5$')
plt.plot(x_values, y_values_5.pdf(x_values), label=r'$\mu=3, \,\;\;\, \sigma^2=0.25$')
plt.xlabel('$x$', size=16)
plt.legend()
plt.savefig('many_normal')
```
